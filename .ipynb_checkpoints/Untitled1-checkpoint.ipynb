```python
# carring out min-max scaling

def min_max_scaling_calculation(feature):
    """
    Performs min-max scaling on the input series.
    Input: series
    Output : scaled series
    """
    scaled = (feature - min(feature)) / (max(feature) - min(feature))
    
    return scaled
```

```python
X_train_numeric_log.apply(min_max_scaling_calculation)
```

### Baseline Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
cross_val_score(model, X_train, y_train, cv = 3)
```

```python
from sklearn.model_selection import cross_validate, ShuffleSplit

splitter = ShuffleSplit(n_splits=3, test_size=0.25, random_state=0)

baseline_scores = cross_validate(
    estimator=model,
    X=X_train,
    y=y_train,
    return_train_score=True,
    cv=splitter
)

print("Train score:     ", baseline_scores["train_score"].mean())
print("Validation score:", baseline_scores["test_score"].mean())
```

```python
y_pred = model.predict(X_test)
residuals1 = y_pred - y_test
```

```python
# Scoring the baseline model
print("MSE baseline: ", mean_squared_error(y_train, model.predict(X_train)))
print("R^2 score baseline: ", r2_score(y_train, model.predict(X_train)))
```
